````python
from sklearn.decomposition import FactorAnalysis

fa = FactorAnalysis(n_components=4)
fa.set_params(rotation='varimax')
fa.fit(df_scaled)
kak=pd.DataFrame(fa.components_, columns=df.columns)
````
